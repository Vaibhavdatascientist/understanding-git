#### Select x and y

```python
x = df.drop('price_range',axis=1)
y = df['price_range']
print(type(x))
print(type(y))
```

#### Split data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#### Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
print(x_train_sc.shape)
print(x_test_sc.shape)
print(type(x_train_sc))
```

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
```

```python
def eval_model(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    print('Training score',model.score(x_train,y_train))
    print('Testing score',model.score(x_test,y_test))
    ypred = model.predict(x_test)
    print('Predictions',ypred)
    cm = confusion_matrix(y_test,ypred)
    print(cm)
    print(classification_report(y_test,ypred))
```

### KNN Classification

```python
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
```

```python
knn = KNeighborsClassifier(n_neighbors = 11)
eval_model(knn,x_train,x_test,y_train,y_test)
```

#### Determine best value for K

```python
test_acc = []

for i in range(3,45,2):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    ypred_knn = model.predict(x_test)
    test_acc.append(model.score(x_test,y_test))
```

```python
plt.figure(figsize=(10,4))
plt.plot(list(range(3,45,2)),test_acc,marker='o')
plt.title('Num of neighbors vs Test Acc')
plt.xlabel('Number of neighbors')
plt.ylabel('Test_Acc')
# plt.legend()
plt.grid()
plt.show()
```
